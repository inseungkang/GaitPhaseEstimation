## Plot Results

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_results = pd.read_csv('../Gait Phase Estimation Results/initial_window_sweep_results.csv', index_col=0)
```

```python
display(df_results)
```

```python
plt.figure(1, figsize=[10,7.5])
plt.scatter(df_results.window_size, df_results.left_rmse_mean)
plt.scatter(df_results.window_size, df_results.right_rmse_mean)
plt.title('LSTM Error vs Window Size')
plt.xlabel('Window Size (datapoints)')
plt.ylabel('RMSE in Gait Phase')
plt.legend(['Left Leg', 'Right Leg'])
plt.show()
```
